```python
# Name: S.Megala
# Roll Number: iitp_aimltn_2602713
# Assignment: Python Loops & Automation - Subjective Question

#----- Task1: Find Maximum and Minimum -----
print("===== Task 1: Find Maximum and Minimum =====")
temperatures = [28, 32, 35, 29, 31, 27, 30]

# Initialize highest and lowest with th first element of the list
highest = temperatures[0]
lowest = temperatures[0]

#use a for loop to check each temperature
for temp in temperatures:
  if temp > highest:
    highest = temp
  if temp < lowest:
    lowest = temp
print(f"Highest Temperature: {highest}°C")
print(f"Lowest Temperature: {lowest}°C")


# --- Task 2: Count Hot Days ---
print("\n===== Task 2: Count Hot Days =====")
temperatures = [28, 32, 35, 29, 31, 27, 30]
hot_days_count = 0

#Use a for loop to iterate through the list
for temp in temperatures:
  if temp <=30:
    continue
  hot_days_count += 1
print(f"Hot Days (>30)°C): {hot_days_count}")


# --- Task 3: Alert System ---
print("\n===== Task 3: Alert System =====")
temperatures = [28, 32, 35, 40, 31, 33, 30]
hot_days_before_alert = 0
day_counter = 1
alert_msg = "" # Variable to store the message for later printing

for temp in temperatures:
    if temp >= 40:
        alert_msg = f"Alert! Extreme temperature {temp}°C detected on Day {day_counter}"
        break

    if temp > 30:
        hot_days_before_alert += 1

    day_counter += 1

print(f"Hot Days before alert: {hot_days_before_alert}")
if alert_msg:
    print(alert_msg)
```
